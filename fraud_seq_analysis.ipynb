```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("data/fraud_planner_v3.csv")
df.head()
```

```python
# Filter successfully generated sequences
successful_df = df[df['sequence'].notna()]
successful_df.head()
```

```python
len(successful_df)
```

```python
plt.hist(successful_df['num_attempts'])
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Distribution of Data")
plt.show()
```

```python
plt.hist(successful_df['per_syntax_errors'])
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Distribution of Data")
plt.show()
```

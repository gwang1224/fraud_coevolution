```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statistics
```

```python
df1 = pd.read_csv("data/fraud_planner_v2.csv")
df1.head()
```

```python
df2 = pd.read_csv("data/fraud_planner_v3.csv")
df2.head()
```

```python
# Filter successfully generated sequences
successful_df1 = df1[df1['sequence'].notna()]
len(successful_df1)
```

```python
successful_df2 = df2[df2['sequence'].notna()]
len(successful_df2)
```

```python
print(statistics.mean(successful_df1['num_attempts']))
print(statistics.mean(df1['num_attempts']))
```

```python
plt.figure(figsize=(6,4))

plt.hist(
    successful_df1['num_attempts'],
    bins=range(1, 12),
    align='left',
    rwidth=0.8,
    color='skyblue',
)

plt.xticks(range(1, 11))

plt.xlim(0.5, 10.5)
plt.xlabel("Attempts")
plt.ylabel("Frequency")
plt.title("Version 2: Attempts taken to generate a valid sequence")
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
print(statistics.mean(successful_df2['num_attempts']))
print(statistics.mean(df2['num_attempts']))
```

```python
plt.figure(figsize=(6,4))

plt.hist(
    successful_df2['num_attempts'],
    bins=range(1, 12),
    align='left',
    rwidth=0.8,
    color='skyblue',
)

plt.xticks(range(1, 11))

plt.xlim(0.5, 10.5)
plt.xlabel("Attempts")
plt.ylabel("Frequency")
plt.title("Version 3: Attempts taken to generate a valid sequence")
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Version 2 errors
print(statistics.mean(df1['per_syntax_errors']))
print(statistics.mean(df1['per_semantic_errors']))
```

```python
# Version 3 errors
print(statistics.mean(df2['per_syntax_errors']))
print(statistics.mean(df2['per_semantic_errors']))
```
